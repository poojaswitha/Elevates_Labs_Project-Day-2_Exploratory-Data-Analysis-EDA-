```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load your dataset
df = pd.read_csv("tested.csv")  # Adjust path if needed
```

```python
# Set plot style
sns.set(style="whitegrid")
```

```python
# Create a 2x2 grid of plots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
```

```python
# 1. Survival count
sns.countplot(x='Survived', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Survival Count')
```

```python
# 2. Passenger class distribution
sns.countplot(x='Pclass', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Passenger Class Distribution')
```

```python
# 3. Age distribution
sns.histplot(df['Age'].dropna(), bins=30, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Age Distribution')
```

```python
# 4. Fare distribution
sns.histplot(df['Fare'].dropna(), bins=30, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Fare Distribution')
```

```python
plt.tight_layout()
plt.show()
```
